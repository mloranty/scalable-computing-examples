# xarray hands-on

```python
import os
import urllib
import pandas as pd
import numpy as np

import xarray as xr   # This is the package we'll explore
```

```python
temp_data = np.array(
    [np.zeros([5,5]),
     np.ones([5,5]),
     #np.ones((5,5)), argument here can be a tuple or a list
     np.ones([5,5])*2]).astype(int)

temp_data
```

```python
dims = ('time', 'lat', 'lon')

coords = {
    'time' : pd.date_range('2022-09-01', '2022-09-03'),
    'lat' : np.arange(70, 20, -10), 
    'lon' : np.arange(60, 110, 10)
}
```

```python
attrs = {
    'title': 'temperature across stations', 
    'standard_name': 'air_temperature', 
    'units': 'degree_c'
}
```

```python
temp = xr.DataArray(data = temp_data,
                    dims = dims, 
                    coords = coords,
                    attrs = attrs)

temp
```

```python
temp.attrs['description'] = 'simple example of xarray'

temp
```

```python
temp.lat.attrs = {
    'description': 'latitude', 
    'standard_name': 'grid_latitude',
    'units': 'degrees_N'
}

temp
```

```python
#temp.loc['2022-09-01',:,:]
temp.sel(time = '2022-09-01', lat = 50)
temp.sel(lon = slice(60,80), time = '2022-09-01')
```

```python
temp.mean(dim = 'time')
temp.mean(dim = ( 'time', 'lat'))
```

```python
temp.to_dataset(name = 'temperature')
```

```python
avg_temp = temp.mean(dim = 'time')

data_vars = {
    'temp': temp, 
    'avg_temp': avg_temp
}

attrs = {
    'title' : 'temperature data'
}

temp_dataset = xr.Dataset(data_vars = data_vars, attrs = attrs)

temp_dataset
```

```python
temp_dataset.plot()
```

```python
fp = os.path.join(os.getcwd(),'temp_dataset.nc')
fp

temp_dataset.to_netcdf(fp)
```

```python
check = xr.open_dataset(fp)
check == temp_dataset # this checks values

xr.open_dataset(fp).equals(temp_dataset) # this evaluates the entire file
```
